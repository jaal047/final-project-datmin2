```python
import pandas as pd
df = pd.read_csv('labelled.csv')
df
```

```python
print(df['label'])
```

```python
print(df['label'][0])
```

```python
#Ubah df['label'] ke df sendiri
label = df['label']

df_final = pd.DataFrame()

for i in range(len(label)):
    l = label[i]
    df1 = pd.DataFrame(eval(l))
    df_final = pd.concat([df_final, df1], axis=0)
    
df_final = df_final.reset_index()
```

```python
#ubah rectanglelabels dari list ke str
df_final['rectanglelabels'] = [''.join(map(str,  l)) for l in df_final['rectanglelabels']]
```

```python
df_final
```

```python
df_dipake = df_final[['x', 'y', 'width', 'height', 'rectanglelabels']]
df_dipake
```

```python
df_dipake.to_csv('labelled_final.csv')
```
